# Classifying the iris dataset

## Loading libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## Loading the dataset

```python
data = pd.read_csv('Iris.csv')
```

```python
data.head(20)
```

```python
del data['Id']
```

```python
data.head(10)
```

```python
print(data.shape) 
```

```python
print(data.describe())
```

## Splitting the data into features and labels

```python
y = data['Species'].values
```

```python
x = data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
features_train,features_test,labels_train,labels_test = train_test_split(x,y,test_size=0.4,random_state=2)
```

## Loading the classifier

```python
clf = RandomForestClassifier()
```

```python
clf.fit(features_train,labels_train)
```

## Predicting the test data

```python
pred = clf.predict(features_test)
```

```python
accuracy = accuracy_score(pred,labels_test)
```

```python
print(accuracy)
```

### Getting an accuracy of 93%. Awesome, Right?

## K-fold cross validation

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
k_fold = KFold(n_splits=10)
```

```python
for train,test in k_fold.split(x):
    clf.fit(x[train],y[train]).score(x[test],y[test])
```

```python
cross_val_score(clf,x,y,cv=k_fold,n_jobs=-1)
```
